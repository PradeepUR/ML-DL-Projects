## Importing the Library

```python
import numpy as np
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv("breast-cancer-wisconsin.data", sep = ',')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
# X.shape
```

```python
print(y)
```

```python
X[X == '?'] = np.nan
```

```python
# X = X.replace(r'^\s*$', np.nan, regex = True)
```

```python
X = X.astype(float)
```

```python
pd.isna(X).sum()
```

```python
# pd.isnull(X).sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(X)
X = imputer.transform(X)
```

## Splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.75, random_state = 0)
```

```python
print(X_train)
```

## Fitting the Logistic Regression model on the training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting the test results

```python
y_pred = classifier.predict(X_test)
```

## Making the confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred)*100)
```

## Caculating the Accuracy Score with k-fold cross validation

```python
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(classifier, X_train, y_train, cv = 10) # here cv is number of folds
print("Accuracy: ",cvs.mean()*100,"%")
print("Standard Deviation: ", cvs.std()*100,"%")
```
